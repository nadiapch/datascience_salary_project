```python
import pandas as pd
df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df1 = df.copy()
```

```python
df1
```

# salary parsing

```python
df1['Job Title'].unique
```

```python
df1['Salary Estimate'].unique
```

```python
df1 = df1[df1['Salary Estimate'] != '-1']
```

```python
df1['Salary Estimate']
```

```python
salary = df1['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
minus_kd = salary.apply(lambda x: x.replace('K', ' ').replace('$', ' '))
```

```python
df1['hourly'] = df1['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df1
```

```python
df1['employer_provided'] = df1['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
min_hr = minus_kd.apply(lambda x: x.lower().replace('per hour', ' ').replace('employer provided salary:', ' '))
```

```python
min_hr
```

```python
df1['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
```

```python
df1
```

```python
df1['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
```

```python
df1['avg_salary'] = (df1.min_salary+df1.max_salary)/2
```

```python
df1
```

# Company name text only

```python
df1['company_txt'] = df1.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3], axis = 1)
```

```python
df1
```

# State field

```python
df1['job_state'] = df1['Location'].apply(lambda x: x.split(',')[1])
```

```python
df1
```

```python
df1.job_state.value_counts()
```

# age of company

```python
df1['same_state'] = df1.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
```

```python
df1
```

```python
df1['age'] = df1.Founded.apply(lambda x: x if x <1 else 2021 - x)
```

```python
df1
```
